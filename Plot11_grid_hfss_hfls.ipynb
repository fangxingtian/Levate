```python
x=hfls[:,60,60].data
y=hfss[:,60,60].data
plt.figure()
plt.subplot(311)
plt.plot(x, y, '.')
m, b = np.polyfit(x, y,1)
plt.plot(x, x*m + b)

plt.subplot(312)
x1=x[:100]
y1=y[:100]
plt.plot(x1, y1, 'r.')
m, b = np.polyfit(x1, y1,1)
plt.plot(x1, x1*m + b)

plt.subplot(313)
x2=x[200:300]
y2=y[200:300]
plt.plot(x2, y2, 'r.')
m, b = np.polyfit(x2, y2,1)
plt.plot(x2, x2*m + b)
plt.savefig('hfls_to_hfss_grid6060_NorESM_monmean.eps',format='eps',dpi=1000)
plt.show()
```
